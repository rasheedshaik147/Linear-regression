```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Race'] = le.fit_transform(df['Race'])
df['Marital Status'] = le.fit_transform(df['Marital Status'])
df['T Stage '] = le.fit_transform(df['T Stage '])
df['N Stage'] = le.fit_transform(df['N Stage'])
df['6th Stage'] = le.fit_transform(df['6th Stage'])
df['differentiate'] = le.fit_transform(df['differentiate'])
df['A Stage'] = le.fit_transform(df['A Stage'])
df['Estrogen Status'] = le.fit_transform(df['Estrogen Status'])
df['Progesterone Status'] = le.fit_transform(df['Progesterone Status'])
df['Status'] = le.fit_transform(df['Status'])
```

```python
X = df.drop(['Status'], axis=1)
y = df['Status']
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()

log_reg.fit(X_train,y_train)
```

```python
y_pred_test = log_reg.predict(X_test)
y_pred_test
```

```python
log_reg.coef_
```

```python
log_reg.intercept_
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, recall_score, precision_score, f1_score
accuracy = accuracy_score(y_test, y_pred_test)
accuracy
```

```python

conf_matrix = confusion_matrix(y_test, y_pred_test)

print('Confusion matrix\n\n', conf_matrix)

print('\nTrue positive = ', conf_matrix[1,1])
print('\nTrue Negative = ', conf_matrix[0,0])
print('\nFalse positive = ', conf_matrix[0,1])
print('\nFalse Negative = ', conf_matrix[1,0])
```

```python
print(classification_report(y_test, y_pred_test))
```
